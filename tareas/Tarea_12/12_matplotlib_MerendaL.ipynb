## Ejercicio 1

Graficar las siguientes funciones en los intervalos dados:

1. $f(x) = \cos(x^2)$ en $[-3\pi,3\pi]$
1. $f(x) = x\sin(1/x)$ en $[-1,1]$ (se necesita un poco de ingeniería para lidiar con las infinitas discontinuidades)
1. $J_n(x)$ con $n=1,2,3,\ldots,5$ en el intervalo $[0,10]$ donde $J_n$ es la función de Bessel del primer tipo.


```python
import math
import numpy as np
import scipy as sp
import scipy.special as sps
import matplotlib as mpl
import matplotlib.pyplot as plt

#Set unicode latex support
mpl.rcParams['text.usetex'] = True
```

```python
#Definimos las funciones
def f1(x):
     return  np.cos(np.square(x))

def f2(x):
    #Para la singularidad en x = 0
    #Calculamos: lim de f2 cuando x -> 0 = 0
    if x == 0: return 0
    return  x * np.sin(1 / x)

def f3(x,n):
    #función de bessel de primer tipo y orden n
    return sps.jv(n,x)
```

```python
#Hacemos el primer grafico de f1(x)

xmin = -np.pi * 3
xmax = np.pi * 3
n_interv = 100000

xs = []
ys = []

#Creamos la serie de datos de f1
for i in range(0,n_interv):
    xi = xmin + i * (xmax - xmin) / n_interv
    xs.append(xi)
    ys.append(f1(xi))

plt.plot(xs,ys)
plt.xlabel('x', color = 'white')
plt.ylabel('f(x)', color = 'white')
plt.title(r"$f(x) = cos (x^{2})$", color = 'white')
plt.axhline(0,color ='black')
plt.axvline(0,color = 'black')
plt.tick_params(axis = 'both', which = 'both', color = 'white', labelcolor = 'white')
plt.grid(b = True)
```

```python
#Hacemos el primer grafico de f2(x)

xmin = -1
xmax = 1
n_interv = 100000

xs = []
ys = []

#Creamos la serie de datos de f1
for i in range(0,n_interv):
    xi = xmin + i * (xmax - xmin) / n_interv
    xs.append(xi)
    ys.append(f2(xi))

plt.plot(xs,ys)
plt.xlabel('x', color = 'white')
plt.ylabel('f(x)', color = 'white')
plt.title(r"$f(x) =x * sin(\frac{1}{x})$", color = 'white')
plt.axhline(0,color ='black')
plt.axvline(0,color = 'black')
plt.tick_params(axis = 'both', which = 'both', color = 'white', labelcolor = 'white')
plt.grid(b = True)
```

```python
#Hacemos el primer grafico de f3(x)

xmin = 0
xmax = 10
n_interv = 100000

xs = []
ys = []
#Creamos la serie de datos de f3
for i in range(0,n_interv):
    xi = xmin + i * (xmax - xmin) / n_interv
    xs.append(xi)

#ploteamos desde n=1 hasta n=5
for n in range(0,10):
    aux = []
    for i in range(0,n_interv):
            aux.append(f3(xs[i],n+1))
    ys.append(aux)
#Ploteamos las funciones de bessel de primer tiṕo y orden n = 1 a n = 5

plt.plot(xs,ys[0],color = 'cornflowerblue',label = 'n=1')
plt.plot(xs,ys[1],color = 'orange',label = 'n=2')
plt.plot(xs,ys[2],color = 'green',label = 'n=3')
plt.plot(xs,ys[3],color = 'red',label = 'n=4')
plt.plot(xs,ys[4],color = 'brown',label = 'n=5')

plt.xlabel('x', color = 'white')
plt.ylabel('f(x)', color = 'white')
plt.title(r"$f(x) = J_{n}(x)$", color = 'white')
plt.axhline(0,color ='black')
plt.axvline(0,color = 'black')
plt.tick_params(axis = 'both', which = 'both', color = 'white', labelcolor = 'white')
plt.grid(b = True)
plt.legend(title = '$J_{n}(x)$', loc = 'best',bbox_to_anchor = (1.0,1.0),framealpha = 1)
```

## Ejercicio 2

Dada la función compleja de variable real:

$\mu(t) = e^{it} + \frac{1}{2}e^{6it} + \frac{i}{3}e^{-14it}$,

graficar la curva paramétrica dada por $(\Re \mu(t),\Im \mu(t))$ con $t\in[0,2\pi]$.

Referencia: Creating symmetry : the artful mathematics of wallpaper patterns, Farris, Frank A, Princeton University Press, 2015

```python
def mu(t):
    z = np.exp(1j * t) + (0.5 * np.exp(6j * t)) + ((1/3) * 1j * np.exp(-14j * t))
    return z.real, z.imag

tmin = 0
tmax = 2 * np.pi
n_interv = 100000

ts = []
reals = []
imags= []

#Creamos la serie de datos de f1
for i in range(0,n_interv):
    ti = tmin + i * (tmax - tmin) / n_interv
    ts.append(ti)
    real, imag = mu(ti)
    reals.append(real)
    imags.append(imag)

plt.plot(reals,imags)
#No me deja usar mathfrak no idea why
plt.xlabel(r'$\mathcal{R}(\mu(t))$', color = 'white')
plt.xticks(np.arange(-2.5,2.5,0.5))
plt.ylabel(r'$\mathcal{I}(\mu(t))$', color = 'white')
plt.title(r"$\mu(t) = e^{it} + \frac{1}{2} e^{6it} + \frac{i}{3} e^{-14it} \quad t\in[0,2\pi]$", color = 'white')
plt.axhline(0,color ='black')
plt.axvline(0,color = 'black')
plt.grid(b=True)
plt.tick_params(axis = 'both', which = 'both', color = 'white', labelcolor = 'white')
a = plt.axis('equal')
```
