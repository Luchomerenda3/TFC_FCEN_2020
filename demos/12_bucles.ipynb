```python
xmin = -3.0
xmax =  7.0
nx = 10000

i = 0
fmin = m.inf
fmax = -m.inf

while (i <= nx):
    x = xmin + (xmax - xmin) * i / nx
    fdex = (x-1)*(x+1)*(x-3)
    if fdex < fmin: fmin = fdex
    if fdex > fmax: fmax = fdex
    i += 1
else:
    print(f"el minimo es {fmin:20.6f}, el máximo es {fmax:20.6f}")
```
